```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('road_accidents_sample.csv')
df.head()
```

```python
df.info()
df.describe()
```

```python
df = df.dropna()
df['Year'] = df['Year'].astype(int)
df.head()
```

```python
acc_by_year = df.groupby('Year')['Total Accidents'].sum().reset_index()
plt.figure(figsize=(10,6))
sns.lineplot(data=acc_by_year, x='Year', y='Total Accidents', marker='o')
plt.title('Total Road Accidents in India (Sample Data)')
plt.ylabel('Total Accidents')
plt.xlabel('Year')
plt.show()
```

```python
top_states = df.groupby('State/UT')['Total Accidents'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=top_states.values, y=top_states.index)
plt.title('Total Accidents by State (Sample Data)')
plt.xlabel('Total Accidents (Aggregate)')
plt.show()
```

```python
killed_by_year = df.groupby('Year')['Persons Killed'].sum().reset_index()
plt.figure(figsize=(10,6))
sns.lineplot(data=killed_by_year, x='Year', y='Persons Killed', marker='o', color='r')
plt.title('Persons Killed Over Years (Sample Data)')
plt.ylabel('Persons Killed')
plt.xlabel('Year')
plt.show()
```

```python
df['Severity Index'] = (df['Persons Killed'] / df['Total Accidents']) * 100
plt.figure(figsize=(12,6))
sns.boxplot(x='State/UT', y='Severity Index', data=df)
plt.xticks(rotation=45)
plt.title('Accident Severity Index by State (Sample Data)')
plt.show()
```
